```python
import numpy as np
import pandas as pd
scroes = np.random.randint(50,100,size=(50,5))
scores_df = pd.DataFrame(scroes,
             columns=['國文','數學','英文','地理','歷史'],
             index=range(1,51))
scores_df
```

```python
#'國文','數學','英文','地理','歷史'

scores_df['總分'] = scores_df['國文'] + scores_df['英文'] + scores_df['數學'] + scores_df['地理'] + scores_df['歷史']
scores_df['平均'] = scores_df['總分'] / 5
scores_df
```

```python
scores_df = pd.DataFrame(scroes,
             columns=['國文','數學','英文','地理','歷史'],
             index=range(1,51))
scores_df
sum_values = scores_df.sum(axis=1)
mean_values = scores_df.mean(axis=1)
```

```python
scores_df['總分'] = sum_values
scores_df['平均'] = mean_values
scores_df
```

```python
scores_df.sort_values(by='總分',ascending=False).iloc[:5]
```

```python
#使用subscript[],過濾資料
scores_df[scores_df['平均'] < 80]
```

```python
#使用subscript[]的搜尋方式
#平均80以下,歷史不及格的
scores_df[(scores_df['平均'] < 80) & (scores_df['歷史'] < 60)]
```

```python
#使用query()的搜尋
scores_df.query('平均<80')
```

```python
scores_df.query('平均<80 and 歷史<60')
```
